```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
#import wandb
#from wandb.keras import WandbCallback
from tensorflow.keras.optimizers import Adam,SGD
from keras.callbacks import optimizer
```

```python
learning_rate_a = 0.01
epochs_a = 25
batch_size_a = 10
```

```python
# Funci√≥n a) 3sin(pix)
x_train_a = np.linspace(-1, 1, 1000)
y_train_a = 3 * np.sin(np.pi * x_train_a)
```

```python
model = keras.Sequential([
    layers.Dense(10, activation='relu', input_shape=(1,)),
    layers.Dense(20, activation='relu'), # Agregar capa intermedia con 64 neuronas
    layers.Dense(1)
])

model.compile(loss='mse', optimizer= SGD(learning_rate=learning_rate_a))

# Entrenar modelo 
history3 = model.fit(x_train_a, y_train_a,
                    batch_size=batch_size_a,
                    epochs=1000,
                    verbose=1)

model.summary()#informacion de la red

# Evaluar modelo
x_test_a = np.linspace(-1, 1, 10)
y_test_a = 3 * np.sin(np.pi * x_test_a)
loss_a3 = model.evaluate(x_test_a, y_test_a, verbose=1)
```

```python
# Graficar funci√≥n original y soluci√≥n de la red
plt.plot(x_test_a, y_test_a, label='Funci√≥n original')
plt.plot(x_test_a, model.predict(x_test_a), label='Red neuronal')
plt.legend()
plt.title('Funci√≥n a) 3sin(pix) red 3')
plt.show()
```

```python
learning_rate_b = 0.01
epochs_b = 500
batch_size_b = 64
```

```python
# Funci√≥n b) 1+2x+4x^3
x_train_b = np.linspace(-1, 1, 1000)
y_train_b = 1 + 2 * x_train_b + 4 * x_train_b**3
```

```python
# Definir modelo
model_b = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(1,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
])
```

```python
# Compilar modelo
model_b.compile(loss='mse', optimizer='adam')

# Entrenar modelo
history_b = model_b.fit(x_train_b, y_train_b,
                    batch_size=64,
                    epochs=500,
                    verbose=1)
```

```python
# Evaluar modelo
x_test_b = np.linspace(-1, 1, 10)
y_test_b = 1 + 2 * x_test_b + 4 * x_test_b**3
loss_b = model_b.evaluate(x_test_b, y_test_b, verbose=1)
```

```python
# Graficar funci√≥n original y soluci√≥n de la red
plt.plot(x_test_b, y_test_b, label='Funci√≥n original')
plt.plot(x_test_b, model_b.predict(x_test_b), label='Red neuronal')
plt.legend()
plt.title('Funci√≥n b) 1+2x+4x^3')
plt.show()
```
